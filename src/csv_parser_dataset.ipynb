```python
def find_brackets_in_list(list_contains_brackets):
  '''

  Buscador de corchetes en lista.

    Parametros : list (lista con componentes String)

    Retorna: list (retorna una lista booleana, true: si se encontró un corchete, false: en caso que no)

  '''
  return [x.find('[')>=0 or x.find(']')>=0 for x in list_contains_brackets]

list_example = ['1', '10', '-9', 'i', '[owr', 'yyy[]', ']utn']
print(list_example)
print(find_brackets_in_list(list_example))
```

```python
def index_brackets_in_list(list_contains_brackets):
  '''

  Buscador de corchetes en lista.

    Parametros : list (lista con componentes String)

    Retorna: list (retorna una lista con los índices de las posiciones donde se encontró 
    un corchete)

  '''
  res = [x.find('[')>=0 or x.find(']')>=0 for x in list_contains_brackets]
  return [i for i, val in enumerate(res) if val]

list_example = ['1', '10', '-9', 'i', '[owr', 'yyy[]', ']utn']
print(list_example)
print(index_brackets_in_list(list_example))
```

```python
def remove_brackets_in_list(list_contains_brackets):
  '''
  
  Removedor de corchetes en lista de Strings.

    Parametros : list (lista con componentes Strings)

    Retorna: list (retorna la lista con los corchetes removidos)

  '''
  aux_list =  [x.replace('[', '') for x in list_contains_brackets]
  return [x.replace(']', '') for x in aux_list]
   

list_example = ['1', '10', '-9', 'i', '[owr', 'yyy[]', ']utn']
print(list_example)
print(remove_brackets_in_list(list_example))
```

```python
def expands_headers_list(list_headers, indexs_to_extends):
  '''
  
  Removedor de corchetes en lista de Strings.

    Parametros :  - list (lista con componentes Strings)
                  - list (lista numérica)

    Retorna: list (retorna la lista extendiendo a un array)

  '''
  new_headers = list_headers.copy() 
  init_header = list_headers[indexs_to_extends[0]]
  new_headers.pop(indexs_to_extends[0])
  for i in range(indexs_to_extends[0], indexs_to_extends[-1]+1):
    new_string = (init_header + '[' + str(i-indexs_to_extends[0]) + ']')
    new_headers.insert(i, new_string) 
  return new_headers
   

list_example = ['1', '10', '-9', 'i', '[owr', 'yyy[]', ']utn']
print(list_example)
print(expands_headers_list(list_example, [3, 5]))
```
